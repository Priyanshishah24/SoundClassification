```python
import pandas as pd
import os
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
dff=pd.read_csv(r'E:\features.csv',index_col=0) #reading features.csv file from file path
dff.head()
```

```python
dff1=dff.rename(columns={dff.columns[4]:'4'}) #renaming unnamed columns
dff1
```

```python
dft=pd.read_csv(r'E:\target.csv',index_col=0) # reading target.csv file 
dft.head()
```

```python
dft1=dft.rename(columns={dft.columns[0]: 'Sounds'}) # renaming the unnamed column
dft1
```

```python
dfm=pd.concat([dft1,dff1],axis=1) # Joining two dataframes to get data from two different dataframes into one dataframe
dfm
```

```python
dfm.duplicated()#checking for duplicate values
```

```python
dfm1=dfm.fillna(value=0)#filling null values with zero
dfm1
```

```python
def createbarchart():
    sns.countplot(data=dfm ,y= 'Sounds')#creating function of createbarchart for bar plot visualization
    plt.show()

```

```python
def createpiechart():#creating function to plot pie chart 
    value_counts_result = dfm1['Sounds'].value_counts()
    value_counts_result
    mylabels=['dog_bark','children_playing','air_conditioner','street_music','engine_idling','jackhammer','drilling','siren','car_horn','gun_shot']             
    plt.figure(figsize=(7, 7))
    plt.pie(value_counts_result,labels=mylabels)
    plt.title('Pie Chart for Sound')
    plt.show()


  
   
        
     
```

```python
dfm2=dfm.describe() # Running describe() function to get descriptive analysis
```

```python
def createheatmap(): # creating heatmap function 
    plt.figure(figsize=(15,10))
    sns.heatmap(dfm2,cmap='RdBu',linewidth=0.5)
    plt.show
    
```

```python
x=dfm2.loc['count':'max','0':'50']#slicing data frame dfm2
x



```

```python
def createboxplot():
    plt.boxplot(x)
    plt.show()
```

```python
createboxplot()# calling createboxplot function
```

```python
def createhistogram():#creating function for histogram
    plt.figure(figsize=(20,10))
    sns.histplot(data=dfm1, x='Sounds')
    plt.show()
    


```

```python
createhistogram() #calling createhistogram function
```
